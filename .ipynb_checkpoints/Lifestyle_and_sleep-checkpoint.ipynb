```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, mean_absolute_error, r2_score
```

```python
# Inspect Dataset for missing or null values
lifestyle_sleep = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
print(lifestyle_sleep.head(10))
print(lifestyle_sleep.nunique())
print(lifestyle_sleep.isnull().sum() / len(lifestyle_sleep))
```

```python
life_sleep_encoded = pd.get_dummies(lifestyle_sleep, columns=['Gender', 'Occupation', 'BMI Category', 'Blood Pressure'], dtype=int)
print(life_sleep_encoded.head(5))
```

```python
# Dropping columns 'Person ID' and 'Sleep Disorder'
columns_to_drop = ['Person ID', 'Sleep Disorder']
cleaned_lifestyle_sleep = life_sleep_encoded.drop(columns=columns_to_drop)
```

```python
# Creating training and test data
y = cleaned_lifestyle_sleep['Quality of Sleep']
X = cleaned_lifestyle_sleep.drop(columns='Quality of Sleep')
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

# Scale the data for better convergence
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Train the baseline LogisticRegression Model
sleep_model_lr = LogisticRegression(max_iter=1000)
sleep_model_lr.fit(x_train_scaled, y_train)
y_pred_lr = sleep_model_lr.predict(x_test_scaled)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
```

```python
print("\n- Logistic Regression Model -")
print(f"Accuracy: {accuracy_lr:.2f}")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
print("Classification Report:")
print(classification_report(y_test, y_pred_lr))
```

```python
# Training RandomForestRegressor (new model)
sleep_model_rf = RandomForestRegressor(n_estimators=100, random_state=0)
sleep_model_rf.fit(x_train_scaled, y_train)
```

```python
# Predict using your new model and print Confusion Matrix and Classification Report
y_pred_rf = sleep_model_rf.predict(x_test_scaled)

print("\n- Random Forest Regressor Model -")

print(f"MAE: {mean_absolute_error(y_test, y_pred_rf):.2f}")

print(f"R-squared: {r2_score(y_test, y_pred_rf):.2f}")

accuracy_rf = accuracy_score(y_test, y_pred_rf.round())
print(f"Accuracy (rounded): {accuracy_rf:.2f}")
```

```python
# Splitting up the training data for the pipeline
lifestyle_sleep[['systolic_bp', 'diastolic_bp']] = lifestyle_sleep['Blood Pressure'].str.split('/', expand=True).astype(int)
columns_to_drop = ['Person ID', 'Sleep Disorder', 'Blood Pressure']
clean_lifestyle_sleep = lifestyle_sleep.drop(columns=columns_to_drop)
y = clean_lifestyle_sleep['Quality of Sleep']
X = clean_lifestyle_sleep.drop(columns='Quality of Sleep')
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
print(clean_lifestyle_sleep.head(5))
```

```python
# Splitting the data into numerical and catergorical variables
num_cols =['Age', 'Sleep Duration', 
'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps', 'systolic_bp', 'diastolic_bp']
cat_cols =['Gender', 'Occupation', 'BMI Category']
```

```python
# Use the ColumnTransformer to preprocess
preprocess  = ColumnTransformer(transformers=[("num", StandardScaler(), num_cols), ("cat", OneHotEncoder(handle_unknown='ignore'),
                                                                                    cat_cols)], remainder='passthrough')
```

```python
#Creating the pipeline
pipeline = Pipeline([("preprocess", preprocess), ('regressor', RandomForestRegressor(n_estimators=100, random_state=0))])
```

```python
#Fitting the training data to the pipeline and print the MAE and R-squared
pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)

print("\n- Random Forest Regressor Pipeline -")
print(f"MAE: {mean_absolute_error(y_test, y_pred):.2f}")
print(f"R-squared: {r2_score(y_test, y_pred):.2f}")
```

```python
# Building Linear Regression pipeline and using it to predict
lr_pipeline = Pipeline(steps=[
    ('preprocess', preprocess),
    ('regressor', LinearRegression())])

lr_pipeline.fit(x_train, y_train)

lr_y_pred = lr_pipeline.predict(x_test)

print("\n- Linear Regressor Pipeline -")
print(f"MAE: {mean_absolute_error(y_test, lr_y_pred):.2f}")
print(f"R-squared: {r2_score(y_test, lr_y_pred):.2f}")
```

```python
#Using GridSearchCV for the best model(using 3-fold cross_validation)
print("\n- Performing Grid Search for Best Random Forest Model -")
param_grid = {'regressor__n_estimators': [50, 100, 150],'regressor__max_depth': [None, 5, 10],
    'regressor__min_samples_leaf': [1, 2]}

gs = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=3, scoring='r2')
gs.fit(x_train, y_train)
```

```python
# Print best estimator and the parameters
```

```python
best_rf_pipeline = gs.best_estimator_
best_params = gs.best_params_
best_score = gs.best_score_

print(f"Best Parameters found: {best_params}")
print(f"Best cross-validation R-squared score: {best_score:.2f}")
```

```python
# Printing the best pipeline, MAE, and R-squared
rf_y_pred = best_rf_pipeline.predict(x_test)

print("\n- Best Random Forest Regressor Pipeline -")
print(f"MAE: {mean_absolute_error(y_test, rf_y_pred):.2f}")
print(f"R-squared: {r2_score(y_test, rf_y_pred):.2f}")
```

```python
# Data Visualization 
print("\n- Data Visualization -")
# Scatterplot of Sleep Duration and Quality of Sleep 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sleep Duration', y='Quality of Sleep', data=lifestyle_sleep)
plt.title('Relationship Between Sleep Duration and Sleep Quality')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Quality of Sleep (1-10)')
plt.grid(True)

plt.savefig('assets/sleep_relationship.png')
plt.show()
```

```python
# Boxplot of Occupation and Quality of Sleep

plt.figure(figsize=(12, 6))
sns.boxplot(x='Occupation', y='Quality of Sleep', data=lifestyle_sleep)
plt.title('Sleep Quality by Occupation')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Occupation')
plt.ylabel('Quality of Sleep (1-10)')
plt.tight_layout()
plt.savefig('assets/sleep_occupations.png')
plt.show()
feature_names = best_rf_pipeline.named_steps['preprocess'].get_feature_names_out()
rf_model = best_rf_pipeline.named_steps['regressor']
importances = rf_model.feature_importances_

# Featured DataFrame Visualization
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance for Quality of Sleep Prediction')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('assets/sleep_features.png')
plt.show()
```

```python
# Saving the best estimator and file
lifestyle_sleep_model = 'rf_sleep_quality_model.joblib'
joblib.dump(best_rf_pipeline, lifestyle_sleep_model)
print(f"\nModel saved to '{lifestyle_sleep_model}'")
```

```python
# Making predictions on new data
# Creating new data point
print("\n- Prediction on New Data -")

brand_new_data = pd.DataFrame([{
    'Age': 34,
    'Gender': 'Female',
    'Occupation': 'Software Engineer',
    'BMI Category': 'Overweight',
    'Sleep Duration': 9.1,
    'Physical Activity Level': 35,
    'Stress Level': 6,
    'Heart Rate': 72,
    'Daily Steps': 12000,
    'systolic_bp': 120,
    'diastolic_bp': 79,
}])

loaded_model = joblib.load(lifestyle_sleep_model)
prediction = loaded_model.predict(brand_new_data)

print(f"New Data Point:\n{brand_new_data}")

print(f"Predicted Quality of Sleep: {prediction[0]:.2f}")
```
